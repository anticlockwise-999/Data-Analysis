# Data Aggregation


**DataFrame.agg(func=function, str, list or dict, axis=0/1)** 

```python
df = pd.DataFrame({'data1' : np.random.randn(5),'data2' : np.random.randn(5),'key1' : ['a', 'a', 'b', 'b', 'a'], 'key2' : ['one', 'two', 'one', 'two', 'one']})
```

```python
df
```

```python
grouped = df.groupby('key1')
```

```python
grouped['data1'].quantile(0.9)
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min() 
```

```python
grouped.agg(peak_to_peak)
```

```python
 grouped.describe()
```

# Column-Wise and Multiple Function Application

```python
tips = pd.read_csv('H:/study/tips.csv')
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips[:6]
```

```python
grouped = tips.groupby(['day', 'smoker'])
```

```python
grouped_pct = grouped['tip_pct']
```

```python
tips['tip_pct'].groupby([tips['day'],tips['smoker']]).mean()
```

```python
#or we can group this way
grouped_pct.agg('mean')
```

```python
grouped_pct.agg(['mean', 'std', peak_to_peak]) #since, .agg only takes function
```

```python
grouped_pct.agg([('foo', 'mean'), ('bar', np.std)]) #rename
```

```python
functions = ['count', 'mean', 'max']
result = grouped['tip_pct', 'total_bill'].agg(functions)
result
```

```python
result['tip_pct']
```

```python
#a list of tuples with custom names can be passed:

ftuples = [('Average', 'mean'), ('Variance', np.var)]

grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
grouped.agg({'tip' : np.max, 'Size' : 'sum'})
```

```python
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],'Size' : 'sum'})
```

# 9.3 General split-apply-combine

**DataFrame.apply(func, axis=0/1, raw=default False, result_type=None, args=(), **kwds)**

```python

#select the top five tip_pct values by group five tip_pct values by group.
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:] 
```

```python
top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top) 
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
```

```python
result
```

```python
result.unstack('smoker')
```

```python
f = lambda x: x.describe()
grouped.apply(f)
```

# Suppressing the Group Keys


In the preceding examples, you see that the resulting object has a hierarchical index
formed from the group keys along with the indexes of each piece of the original
object. You can disable this by passing **group_keys=False** to groupby:

```python
tips.groupby('smoker', group_keys=False).apply(top)
```

# Quantile and Bucket Analysis

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),'data2': np.random.randn(1000)})
```

```python
quartiles = pd.cut(frame.data1, 4)
```

```python
quartiles[:10]
```

```python
 def get_stats(group):
        return {'min': group.min(), 'max': group.max(),
                'count': group.count(), 'mean': group.mean()}
```

```python
grouped = frame.data2.groupby(quartiles)
```

```python
grouped.apply(get_stats).unstack()
```

# Example: Filling Missing Values with Group-Specific Values

```python
s = pd.Series(np.random.randn(6))
```

```python
s[::2] = np.nan
```

```python
s
```

```python
s.fillna(s.mean())
```

```python
states = ['Ohio', 'New York', 'Vermont', 'Florida','Oregon', 'Nevada', 'California', 'Idaho']
```

```python
group_key = ['East'] * 4 + ['West'] * 4
```

```python
data = pd.Series(np.random.randn(8), index=states)
```

```python
data
```

```python
data[['Vermont', 'Nevada', 'Idaho']] = np.nan
```

```python
data
```

```python
data.groupby(group_key).mean() 
```

```python
fill_mean = lambda g: g.fillna(g.mean()) #the function
```

```python
data.groupby(group_key).apply(fill_mean) #filling mean using group_key #.apply(function)
```

```python
fill_values = {'East': 0.5, 'West': -1}
```

```python
fill_func = lambda g: g.fillna(fill_values[g.name])
```

```python
data.groupby(group_key).apply(fill_func)
```
