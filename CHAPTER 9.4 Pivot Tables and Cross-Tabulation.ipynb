```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('H:/study/tips.csv')
```

```python
tips.head(3)
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

**pandas.pivot_table(data, values=None, index=None, columns=None, aggfunc=(default mean)function/ list of functions/ dict, fill_value=None, margins=False, dropna=True, margins_name='All', observed=False)**

```python
tips.pivot_table(index=['day', 'smoker']) #mean values
```

```python
tips.pivot_table(['tip_pct', 'Size'], index=['time', 'day'],columns='smoker')
```

```python
tips.pivot_table(['tip_pct', 'Size'], index=['time', 'day'],columns='smoker', margins=True)
```

```python
tips.pivot_table('tip_pct', index=['time', 'smoker'], columns='day',aggfunc=len, margins=True)
```

```python
tips.pivot_table('tip_pct', index=['time', 'Size', 'smoker'],columns='day', aggfunc='mean', fill_value=0)
```
